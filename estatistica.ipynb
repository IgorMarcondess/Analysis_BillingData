```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {
    'idade': [15, 18, 25, 25, 40, 55, 58, 60, 80], 
    'altura': [160, 162, 165, 168, 172, 174, 174, 174, 176]
}
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
df_medidas
```

### Medidas de Posição

```python
df_medidas["idade"].mean() #media
```

```python
df_medidas["idade"].median() #mediana
```

```python
df_medidas["idade"].mode() #moda
```

### Medidas de Dispersão

```python
df_medidas.idade.var() #Variância
```

```python
df_medidas.idade.std() #Desvio padrao
```

```python
# Coeficiente de Variação
# IDADE
coeficiente = df_medidas.idade.std() / df_medidas.idade.mean() * 100
coeficiente
```

```python
# ALTURA
coeficiente = df_medidas.altura.std() / df_medidas.altura.mean() * 100
coeficiente
```

### Medidas de Forma

```python
# Assimetria - Idade
df_medidas.idade.skew()
```

```python
# Assimetria - Altura
df_medidas.altura.skew()
```

```python
# Curtose - Idade
df_medidas.idade.kurtosis()
```

```python
# Curtose - Altura
df_medidas.altura.kurtosis()
```

Curtose alto == Dados mais concentrados


Curtose baixo ou negativo == Dados masi espalhados

### Medidas de Correlação

```python
#Metodo de Pearson
df_medidas.corr(method='pearson')
```

```python
#Metodo de Spearman
df_medidas.corr(method='spearman')
```

```python
#Utilizando os metodos definindo as variáveis de forma Explícita
df_medidas.idade.corr(df_medidas.altura)
```

#### Gráfico de Barras

```python
#Preparando os dados para Gráfico de barras - pois é necessário uma medida qualitativo e uma medida numérica
dict_vendas = {'categoria': ['masculino', 'feminino', 'infantil', 'casa'], 'valor':
[400000, 600000, 250000, 580000], 'quantidade': [3000, 5000, 1500, 2500]}
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
df_vendas
```

```python
#Grafico de barras - Vertical
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
#Grafico de barras - Vertical Ordenado
df_vendas.sort_values('valor').plot.bar(x='categoria', y='valor')
```

```python
#Grafico de barras - Horizontal
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
#Grafico de barras - Horizontal Ordenado
df_vendas.sort_values('quantidade').plot.barh(x='categoria', y='quantidade')
```
