```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Com base nos dados abaixo, devemos responder essas perguntas:
Qual a média de vendas?

Crie um gráfico vertical e um gráfico de linhas mostrando o mês de referência

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

### Criando o DataFrame

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

```python
df_faturamento.shape
```

##### A média dos valores é de R$6.420.83

```python
df_faturamento['valor'].mean()
```

##### Aqui entendemos que Mês de referência é Dezembro/2023

```python
df_faturamento.sort_values('valor').plot.bar(x='data_ref', y='valor')
```

##### No gráfico de linha abaixo, conseguimos idnetificar que a data Dezembro/2023 também é o Mês referência

```python
valores = df_faturamento['valor']
meses = df_faturamento['data_ref']
plt.figure(figsize=(8, 6))
plt.plot(meses, valores, label="Data")
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
